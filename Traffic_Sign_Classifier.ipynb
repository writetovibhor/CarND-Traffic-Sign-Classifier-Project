### Pre-process the Data Set (normalization, grayscale, etc.)

Use the code cell (or multiple code cells, if necessary) to implement the first step of your project.

````python
### Preprocess the data here. Preprocessing steps could include normalization, converting to grayscale, etc.
### Feel free to use as many code cells as needed.
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
import random
# Visualizations will be shown in the notebook.
%matplotlib inline

DO_TRAIN = False
DO_PLOT = False

if DO_TRAIN:
    OFFSET = 5
    image = X_train[0].squeeze()
    scale = 1 + (OFFSET * 2) / image.shape[0]
    r1 = np.float32([[0,0],[image.shape[1],0],[image.shape[1],image.shape[0]],[0,image.shape[0]]])
    r2 = np.float32([[0,0],[image.shape[0] - OFFSET,OFFSET],[image.shape[1]-OFFSET,image.shape[0]-OFFSET],[0,image.shape[0]]])
    l1 = np.float32([[0,0],[image.shape[1],0],[image.shape[1],image.shape[0]],[0,image.shape[0]]])
    l2 = np.float32([[OFFSET,OFFSET],[image.shape[1],0],[image.shape[1],image.shape[0]],[OFFSET,image.shape[0]-OFFSET]])
    Mr = cv2.getPerspectiveTransform(r1,r2)
    Ml = cv2.getPerspectiveTransform(l1,l2)

    def changePerspective(frame, edge):
        res = cv2.resize(frame,None,fx=scale, fy=scale, interpolation = cv2.INTER_CUBIC)
        if edge == 'right':
            changed = cv2.warpPerspective(res,Mr,(res.shape[1], res.shape[0]))
            y_offset = int((changed.shape[0] - frame.shape[0]) * 0.5)
            return changed[y_offset:y_offset + frame.shape[0], 0:frame.shape[1]]
        elif edge == 'left':
            changed = cv2.warpPerspective(res,Ml,(res.shape[1],res.shape[0]))
            y_offset = int((changed.shape[0] - frame.shape[0]) * 0.5)
            return changed[y_offset:y_offset + frame.shape[0], changed.shape[1] - frame.shape[1]:changed.shape[1]]

    def equalize(frame):
        yuv = cv2.cvtColor(frame, cv2.COLOR_RGB2YUV)
        yuv[:,:,0] = cv2.equalizeHist(yuv[:,:,0])
        return cv2.cvtColor(yuv, cv2.COLOR_YUV2RGB)

    def grayscale(frame):
        gray = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
        gray3ch = cv2.cvtColor(gray, cv2.COLOR_GRAY2RGB)
        return gray3ch


    # Add histogram equalized images
    X_train = np.concatenate((X_train, np.array([equalize(x) for x in X_train])), axis=0)
    y_train = np.concatenate((y_train, y_train), axis=0)

    # Add grayscale images
    X_train = np.concatenate((X_train, np.array([grayscale(x) for x in X_train])), axis=0)
    y_train = np.concatenate((y_train, y_train), axis=0)

    # Add perspective change

    X_train = np.concatenate((X_train, np.array([changePerspective(x, 'right') for x in X_train]), np.array([changePerspective(x, 'left') for x in X_train])), axis=0)
    y_train = np.concatenate((y_train, y_train, y_train), axis=0)

    # Plot
    random_index = random.randint(0, n_train)
    print("random_index = {}, len(X_train) = {}".format(random_index, len(X_train)))

    if DO_PLOT:
        fig = plt.figure()
        def plotImage(index, position):
            image = X_train[index].squeeze()
            fig.add_subplot(4,4,position)
            plt.imshow(image)
        # Original input
        plotImage(random_index, 1)
        # Histogram equalized
        plotImage(random_index + n_train * 1, 2)
        # Grayscale
        plotImage(random_index + n_train * 2, 3)
        # Grayscale equalized
        plotImage(random_index + n_train * 3, 4)
        # Original right perspective
        plotImage(random_index + n_train * 4, 5)
        # Equalized right perspective
        plotImage(random_index + n_train * 5, 6)
        # Grayscale right perspective
        plotImage(random_index + n_train * 6, 7)
        # Grayscale equalized right perspective
        plotImage(random_index + n_train * 7, 8)
        # Original left perspective
        plotImage(random_index + n_train * 8, 9)
        # Equalized left perspective
        plotImage(random_index + n_train * 9, 10)
        # Grayscale left perspective
        plotImage(random_index + n_train * 10, 11)
        # Grayscale left right perspective
        plotImage(random_index + n_train * 11, 12)

# Normalize
X_train = (X_train - 128.0) / 128.0
X_valid = (X_valid - 128.0) / 128.0
X_test = (X_test - 128.0) / 128.0

print(X_train.shape)
print(y_train.shape)
````

### Model Architecture

````python
### Define your architecture here.
### Feel free to use as many code cells as needed.
import tensorflow as tf

print("tensorflow version: ", tf.__version__)
EPOCHS = 25
BATCH_SIZE = 256
keep_prob = tf.placeholder(tf.float32)
````

````python
from tensorflow.contrib.layers import flatten
from tensorflow.contrib.layers import xavier_initializer
def LeNet(x):    
    class Specs:
        class Conv1:
            Filter = [5, 5, 3, 18]
            Stride = [1, 1, 1, 1]
            Padding = 'VALID'
            Bias = Filter[3]
        class Pool1:
            Kernel_Size = [1, 2, 2, 1]
            Stride = [1, 2, 2, 1]
            Padding = 'VALID'
        class Conv2:
            Filter = [5, 5, 18, 48]
            Stride = [1, 1, 1, 1]
            Padding = 'VALID'
            Bias = Filter[3]
        class Pool2:
            Kernel_Size = [1, 2, 2, 1]
            Stride = [1, 2, 2, 1]
            Padding = 'VALID'
        class Dense1:
            Units = 360
        class Dense2:
            Units = 252
        class Dense3:
            Units = n_classes

    # TODO: Layer 1: Convolutional. Input = 32x32x3. Output = 28x28x18.
    #filter1 = tf.Variable(tf.truncated_normal(Specs.Conv1.Filter, mean=0, stddev=0.1), name='filter1')
    filter1 = tf.get_variable(name='filter1', shape=Specs.Conv1.Filter, initializer=xavier_initializer())
    bias1 = tf.Variable(tf.zeros(Specs.Conv1.Bias), name='bias1')
    conv1 = tf.nn.conv2d(x, filter1, Specs.Conv1.Stride, Specs.Conv1.Padding, name='conv1') + bias1
    norm1 = tf.nn.l2_normalize(conv1, 2, name='norm1')
    
    # TODO: Activation.
    relu1 = tf.nn.relu(norm1, name='relu1')

    # TODO: Pooling. Input = 28x28x18. Output = 14x14x18.
    pool1 = tf.nn.max_pool(relu1, Specs.Pool1.Kernel_Size, Specs.Pool1.Stride, Specs.Pool1.Padding, name='pool1')

    # TODO: Layer 2: Convolutional. Output = 10x10x48.
    #filter2 = tf.Variable(tf.truncated_normal(Specs.Conv2.Filter, mean=0, stddev=0.1), name='filter2')
    filter2 = tf.get_variable(name='filter2', shape=Specs.Conv2.Filter, initializer=xavier_initializer())
    bias2 = tf.Variable(tf.zeros(Specs.Conv2.Bias), name='bias2')
    conv2 = tf.nn.conv2d(pool1, filter2, Specs.Conv2.Stride, Specs.Conv2.Padding, name='conv2') + bias2
    norm2 = tf.nn.l2_normalize(conv2, 2, name='norm2')
    
    # TODO: Activation.
    relu2 = tf.nn.relu(norm2, name='relu2')

    # TODO: Pooling. Input = 10x10x48. Output = 5x5x48.
    pool2 = tf.nn.max_pool(relu2, Specs.Pool2.Kernel_Size, Specs.Pool2.Stride, Specs.Pool2.Padding, name='pool2')

    # TODO: Flatten. Input = 5x5x48. Output = 1200.
    flat1 = flatten(pool2)
    
    # TODO: Layer 3: Fully Connected. Input = 1200. Output = 360.
    dense1 = tf.layers.dense(inputs=flat1, units=Specs.Dense1.Units, name='dense1')
    
    # TODO: Activation.
    relu3 = tf.nn.relu(dense1, name='relu3')
    
    # Dropout
    dropout1 = tf.nn.dropout(relu3, keep_prob, name='dropout1')

    # TODO: Layer 4: Fully Connected. Input = 360. Output = 252.
    dense2 = tf.layers.dense(inputs=dropout1, units=Specs.Dense2.Units, name='dense2')

    # TODO: Activation.
    relu4 = tf.nn.relu(dense2, name='relu4')

    # TODO: Layer 5: Fully Connected. Input = 252. Output = 43.
    logits = tf.layers.dense(inputs=relu4, units=Specs.Dense3.Units, name='dense3')
    
    return logits
````

````python
x = tf.placeholder(tf.float32, (None, 32, 32, 3))
y = tf.placeholder(tf.int32, (None))
one_hot_y = tf.one_hot(y, n_classes)
````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
### Train your model here.
### Calculate and report the accuracy on the training and validation set.
### Once a final model architecture is selected, 
### the accuracy on the test set should be calculated and reported as well.
### Feel free to use as many code cells as needed.

rate = 1e-4

logits = LeNet(x)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits=logits, labels=one_hot_y)
loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate = rate)
training_operation = optimizer.minimize(loss_operation)
````

````python
correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()

def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y, keep_prob: 1.0})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples
````

````python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    if DO_TRAIN:
        num_examples = len(X_train)    
        print("Training...")
        print()
        for i in range(EPOCHS):
            X_train, y_train = shuffle(X_train, y_train)
            for offset in range(0, num_examples, BATCH_SIZE):
                end = offset + BATCH_SIZE
                batch_x, batch_y = X_train[offset:end], y_train[offset:end]
                sess.run(training_operation, feed_dict={x: batch_x, y: batch_y, keep_prob: 0.7})
            validation_accuracy = evaluate(X_valid, y_valid)
            print("EPOCH {} ...".format(i+1))
            print("Validation Accuracy = {:.3f}".format(validation_accuracy))
            print()
        saver.save(sess, './lenet')
        print("Model saved")
    else:
        saver.restore(sess, './lenet')
        print("Model restored from ./lenet")
        validation_accuracy = evaluate(X_valid, y_valid)
        print("Validation Accuracy = {:.3f}".format(validation_accuracy))
        
````
